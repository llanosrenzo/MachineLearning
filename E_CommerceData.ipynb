<div style="width: 90%; background-color: #ffffff; color: #ffffff; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px;border-radius:10px;">
    <span style="color: #0057e7;">
        Inicialmente, determinar√© el porcentaje de valores faltantes presentes en cada columna, seguido de la selecci√≥n de la estrategia m√°s efectiva para abordarlos.
    </span>

```python
def report_missings(data):
    # Calcular el porcentaje y la cantidad de valores faltantes
    porc_valores_perdidos = round((data.isnull().sum(axis=0).sort_values(ascending=False) / len(data)) * 100, 3)
    cant_valores_perdidos = data.isnull().sum(axis=0).sort_values(ascending=False)
    
    # Crear un DataFrame para los valores perdidos
    df_valores_perdidos = pd.concat([cant_valores_perdidos, porc_valores_perdidos], axis='columns').reset_index()
    df_valores_perdidos.columns = ['Variables', 'Cant. Nulos', '% Nulos']
    df_valores_perdidos['Cant. No Nulos'] = len(data) - df_valores_perdidos['Cant. Nulos']
    
    # Imprimir la descripci√≥n de los valores perdidos
    print('=' * 100)
    print('*' * 32, 'Descripci√≥n de Valores Perdidos', '*' * 32)
    print('=' * 100)
    print('\n', df_valores_perdidos, '\n')
    
    # Crear un DataFrame para la visualizaci√≥n
    df_perd = pd.DataFrame(df_valores_perdidos, columns=['Variables', 'Cant. Nulos', '% Nulos', 'Cant. No Nulos'])
    
    # Crear los subgr√°ficos
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(25, 17))

    # Gr√°fico de barras del porcentaje de datos faltantes
    sns.barplot(data=df_perd, x='Variables', y='% Nulos', ax=ax1)
    ax1.set_ylabel('Porcentaje de datos faltantes (%)')
    ax1.set_ylim(0, 100)
    ax1.set_title('Porcentaje de datos faltantes por variable', fontweight='bold')
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')

    # Gr√°fico de calor de la distribuci√≥n de valores perdidos
    sns.heatmap(data.isnull().T, ax=ax2, cmap='viridis', cbar=False)
    ax2.set_title('Distribuci√≥n de valores perdidos', fontweight='bold')

    # Gr√°fico de calor de la correlaci√≥n de nulidad entre variables
    missingno.heatmap(data, ax=ax3, cmap='cividis', vmin=0, fontsize=12)
    ax3.set_title('Correlaci√≥n de nulidad entre variables', fontweight='bold')

    # Dendograma basado en la correlaci√≥n de valores faltantes
    missingno.dendrogram(data, ax=ax4, fontsize=12)
    ax4.set_title('Dendograma de correlaci√≥n de valores faltantes', fontweight='bold')
    ax4.set_ylim([0, 1.1])
    ax4.grid(True)

    # Ajustar el dise√±o y aumentar la distancia entre los gr√°ficos
    plt.subplots_adjust(wspace=1.2, hspace=1)  # Ajustar el espacio horizontal y vertical
    plt.tight_layout()
    plt.show()
  
```

```python
try :
    report_missings(df)
except Exception as e:
    pass
```

Los clientes que tienen un valor monetario m√°s alto han contribuido m√°s al negocio, indicando su potencial alto valor a lo largo de la vida.</li>

<a id="recency"></a>
## <b><span style='color:#fcc36d'>Paso 4.1.1 |</span><span style='color:#ff6200'> Recencia (R)</span></b>
‚¨ÜÔ∏è [Tabel of Contents](#contents_tabel)

<div style="width: 90%; background-color: #ffffff; color: #ffffff; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px;border-radius:10px;">
<ul style="list-style-type: none; padding-left: 0; color: #0057e7;">
En este paso, nos enfocamos en entender cu√°n recientemente ha realizado una compra un cliente. Este es un aspecto crucial de la segmentaci√≥n de clientes, ya que ayuda a identificar el nivel de compromiso de los clientes. Aqu√≠, voy a definir la siguiente caracter√≠stica:
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> D√≠as Desde la √öltima Compra: </strong> Esta caracter√≠stica representa el n√∫mero de d√≠as que han pasado desde la √∫ltima compra del cliente. Un valor m√°s bajo indica que el cliente ha comprado recientemente, lo que implica un mayor nivel de compromiso con el negocio, mientras que un valor m√°s alto puede indicar un lapsus o disminuci√≥n del compromiso. Al comprender la recencia de las compras, las empresas pueden adaptar sus estrategias de marketing para reenganchar a los clientes que no han realizado compras en un tiempo, lo que potencialmente aumenta la retenci√≥n de clientes y fomenta la lealtad.</li>

</div>

```python
# Convertir InvoiceDate a datetime
# =================================================================================
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Ahora extraer de InvoiceDate solo la fecha
# =================================================================================
df['InvoiceDay'] = df['InvoiceDate'].dt.date

# Determinar la fecha maxima de compra por cliente
# =================================================================================
cliente_data = df.groupby(by='CustomerID')['InvoiceDay'].max().reset_index()

# Encontrar la fecha m√°s reciente en todo el conjunto de datos
# ==================================================================================
fecha_mas_reciente = df['InvoiceDay'].max()

# Convertir InvoiceDay para tipo datetime antes de la sustracci√≥n
# ==================================================================================
cliente_data['InvoiceDay'] = pd.to_datetime(cliente_data['InvoiceDay'])
fecha_mas_reciente = pd.to_datetime(fecha_mas_reciente)

# Calcular el n√∫mero de d√≠as desde la √∫ltima compra para cada mes
# ==================================================================================
cliente_data['dias_desde_ultima_compra'] = (fecha_mas_reciente - cliente_data['InvoiceDay']).dt.days

# Eliminar la columna InvoiceDay
# ==================================================================================
cliente_data.drop(columns=['InvoiceDay'],inplace=True)
```

<a id="frequency"></a>
## <b><span style='color:#fcc36d'>Paso 4.1.2 |</span><span style='color:#ff6200'> Frecuencia (F)</span></b>
‚¨ÜÔ∏è [Tabel of Contents](#contents_tabel)

<div style="width: 90%; background-color: #ffffff; color: #ffffff; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px;border-radius:10px;">
<ul style="list-style-type: none; padding-left: 0; color: #0057e7;">
En este paso, voy a crear dos caracter√≠sticas que cuantifican la frecuencia del compromiso de un cliente con el minorista:
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> Total de Transacciones:</strong> Esta caracter√≠stica representa el n√∫mero total de transacciones realizadas por un cliente. Ayuda a comprender el nivel de compromiso de un cliente con el minorista.</li>
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> Total de Productos Comprados: </strong> Esta caracter√≠stica indica el total de productos (suma de cantidades) comprados por un cliente en todas las transacciones. Proporciona una visi√≥n del comportamiento de compra del cliente en t√©rminos del volumen de productos adquiridos.</li>

Estas caracter√≠sticas ser√°n cruciales para segmentar a los clientes seg√∫n su frecuencia de compra, lo que es un aspecto clave para determinar segmentos de clientes para marketing dirigido y recomendaciones personalizadas.
</div>

```python
# Calcular el n√∫mero de transacciones  por cada cliente
# ==================================================================================
transacciones = df.groupby(by= 'CustomerID')['InvoiceNo'].nunique().reset_index()
transacciones.rename(columns = {'InvoiceNo':'Total_transacciones'},inplace=True)

# Calcular el total de productos comprados por cada cliente
# ==================================================================================
productos_comprados_totales = df.groupby(by='CustomerID')['Quantity'].sum().reset_index()
productos_comprados_totales.rename(columns = {'Quantity':'Total_productos_comprados'},inplace=True)

# Ahora juntamos ambas variables al cliente_fecha_maxima_compra
# ==================================================================================
cliente_data = pd.merge(cliente_data,transacciones,on='CustomerID')
cliente_data = pd.merge(cliente_data,productos_comprados_totales, on = 'CustomerID')

# Mostrar los resultados del merge
# ==================================================================================
cliente_data.head()
```

<a id="monetary"></a>
## <b><span style='color:#fcc36d'> Paso 4.1.3 |</span><span style='color:#ff6200'> Monetario (M)</span></b>
‚¨ÜÔ∏è [Tabel of Contents](#contents_tabel)

<div style="width: 90%; background-color: #ffffff; color: #ffffff; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px;border-radius:10px;">
<ul style="list-style-type: none; padding-left: 0; color: #0057e7;">
En este paso, voy a crear dos caracter√≠sticas que representan el aspecto monetario de las transacciones de los clientes:
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> Gasto Total:</strong> Esta caracter√≠stica representa la cantidad total de dinero gastado por cada cliente. Se calcula como la suma del producto de <code> UnitPrice</code> y <code> Quantity</code> para todas las transacciones realizadas por un cliente. Esta caracter√≠stica es crucial, ya que ayuda a identificar los ingresos totales generados por cada cliente, lo que es un indicador directo del valor de un cliente para el negocio.</li>
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> Valor Promedio de Transacci√≥n: </strong> Esta caracter√≠stica se calcula como el <code>Gasto Total</code> dividido por el <code>Total de Transacciones</code> de cada cliente. Indica el valor promedio de una transacci√≥n realizada por un cliente. Esta m√©trica es √∫til para comprender el comportamiento de gasto de los clientes por transacci√≥n, lo que puede ayudar a adaptar estrategias de marketing y ofertas a diferentes segmentos de clientes seg√∫n sus patrones de gasto promedio.</li>

</div>

```python
# Calcular el total de gastos por cliente
# ==================================================================================
df['total_gastos'] = df['UnitPrice']*df['Quantity']
total_gastos = df.groupby('CustomerID')['total_gastos'].sum().reset_index()

# Calcular el valor promedio de transacci√≥n para cada cliente.
# ==================================================================================
promedio_transacciones = total_gastos.merge(transacciones,on ='CustomerID')
promedio_transacciones['promedio_transacciones'] = promedio_transacciones['total_gastos']/promedio_transacciones['Total_transacciones']

# Ahora uniendo las nuevas caracteristicas en el cliente_fecha_maxima_compra
# ==================================================================================
cliente_data = pd.merge(cliente_data,total_gastos,on = 'CustomerID')
cliente_data = pd.merge(cliente_data,promedio_transacciones[['CustomerID','promedio_transacciones']],on ='CustomerID')

# Mostrando los resultados
# ==================================================================================
cliente_data.head()
```

<a id="behaviroal_features"></a>
# <b><span style='color:#fcc36d'> Paso 4.3 |</span><span style='color:#ff6200'> Caracter√≠sticas Comportamentales</span></b>
‚¨ÜÔ∏è [Tabla de Contenidos](#contents_tabel)

<div style="width: 90%; background-color: #ffffff; color: #ffffff; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px;border-radius:10px;">
<ul style="list-style-type: none; padding-left: 0; color: #0057e7;">
En este paso, nuestro objetivo es entender y capturar los patrones y comportamientos de compra de los clientes. Estas caracter√≠sticas nos brindar√°n informaci√≥n sobre las preferencias de los clientes en cuanto a cu√°ndo les gusta comprar, lo cual puede ser informaci√≥n crucial para personalizar su experiencia de compra. Aqu√≠ est√°n las caracter√≠sticas que planeo introducir:
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> D√≠as Promedio Entre Compras:</strong> Esta caracter√≠stica representa el n√∫mero promedio de d√≠as que un cliente espera antes de realizar otra compra. Comprender esto puede ayudar a predecir cu√°ndo es probable que el cliente realice su pr√≥xima compra, lo que puede ser una m√©trica crucial para el marketing dirigido y promociones personalizadas.</li>
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> D√≠a Favorito de Compra:</strong> Esto denota el d√≠a de la semana en el que el cliente compra m√°s. Esta informaci√≥n puede ayudar a identificar los d√≠as de compra preferidos de diferentes segmentos de clientes, lo que se puede utilizar para optimizar estrategias de marketing y promociones para diferentes d√≠as de la semana.</li>
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> Hora Favorita de Compra:</strong> Esto se refiere a la hora del d√≠a en la que el cliente compra m√°s. Identificar la hora favorita de compra puede ayudar a optimizar el momento de las campa√±as de marketing y promociones para alinearlas con los momentos en que diferentes segmentos de clientes est√°n m√°s activos.</li>

</div>

```python
# Extrayendo el dia de semana y hora de InvoiceDate
# ==================================================================================
df['Dia_semana'] = df['InvoiceDate'].dt.dayofweek
df['Hora'] = df['InvoiceDate'].dt.hour

# Calcular el n√∫mero promedio de d√≠as entre compras consecutivas
# ==================================================================================
dias_entre_compras = df.groupby(by = 'CustomerID')['InvoiceDay'].apply(lambda x: (x.diff().dropna()).apply(lambda y: y.days))
promedio_dias_entre_compras  = dias_entre_compras.groupby(by = 'CustomerID').mean().reset_index()
promedio_dias_entre_compras.rename(columns = {'InvoiceDay':'Promedio_dias_entre_compras'},inplace=True)

# Calcular el dia favorito de compra de la semana
# ==================================================================================
dia_favorito_compra = df.groupby(['CustomerID','Dia_semana'])['InvoiceNo'].count().reset_index()
dia_favorito_compra.rename(columns={'InvoiceNo':'Count'},inplace=True)
dia_favorito_compra = dia_favorito_compra.loc[dia_favorito_compra.groupby('CustomerID')['Count'].idxmax()][['CustomerID','Dia_semana']]

# Calcular el dia favorito de compra de la semana
# ==================================================================================
Hora_favorito_compra = df.groupby(['CustomerID','Hora'])['InvoiceNo'].count().reset_index()
Hora_favorito_compra.rename(columns= {'InvoiceNo':'Count'},inplace= True)
Hora_favorito_compra = Hora_favorito_compra.loc[Hora_favorito_compra.groupby('CustomerID')['Count'].idxmax()][['CustomerID','Hora']]

# Ahora uniendo las nuevas caracteristicas en el cliente_data
# ==================================================================================
cliente_data =  pd.merge(cliente_data,promedio_dias_entre_compras,on ='CustomerID')
cliente_data =  pd.merge(cliente_data,dia_favorito_compra,on = 'CustomerID')
cliente_data =  pd.merge(cliente_data,Hora_favorito_compra,on ='CustomerID')

# Mostrando el resultado con la nueva variable de productos unicos comprados
# ==================================================================================
cliente_data.head()
```

<ul style="list-style-type: none; padding-left: 0; color: #0057e7;">
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> Primero, agrupar√© los datos por <code>CustomerID</code> y <code>Country</code> y calcular√© el n√∫mero de transacciones por pa√≠s para cada cliente.</li>
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> A continuaci√≥n, identificar√© el pa√≠s principal de cada cliente (el pa√≠s desde el cual tienen el mayor n√∫mero de transacciones).</li>
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> Luego, crear√© una columna binaria que indique si el cliente es del UK o no.</li>
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> Finalmente, fusionar√© esta informaci√≥n con el dataframe cliente_data para incluir la nueva caracter√≠stica en nuestro an√°lisis.</li>

```python
# Agrupar por CustomerID y Country para obtener el n√∫mero de transacciones por pais para cada cliente
# ==================================================================================
cliente_pais = df.groupby(by = ['CustomerID','Country'])['InvoiceNo'].count().reset_index()
cliente_pais.rename(columns = {'InvoiceNo':'Numero_transacciones'},inplace = True)

# Obtener el pa√≠s con el mayor n√∫mero de transacciones para cada cliente (en caso de que un cliente tenga transacciones de m√∫ltiples pa√≠ses)
# ==================================================================================
cliente_pais_principal = cliente_pais.sort_values('Numero_transacciones',ascending = False).drop_duplicates('CustomerID')

# Crear una columna binaria que indique si el cliente es del United Kingdom o no
# ==================================================================================
cliente_pais_principal['IS_UK'] = cliente_pais_principal['Country'].apply(lambda x : 1 if x == 'United Kingdom' else 0)

# Ahora uniendo las nuevas caracteristicas en el cliente_data
# ==================================================================================
cliente_data = pd.merge(cliente_data,cliente_pais_principal[['CustomerID','IS_UK']],on = 'CustomerID',how = 'left')

# Mostrando el resultado con la nueva variable binaria si el cliente es del United Kingdom
# ==================================================================================
cliente_data.head()
```

# <b><span style='color:#fcc36d'>Paso 4.6 |</span><span style='color:#ff6200'> Estacionalidad y Tendencias</span></b>
‚¨ÜÔ∏è [Tabla de Contenidos](#contents_tabel)

<div style="width: 90%; background-color: #ffffff; color: #ffffff; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px;border-radius:10px;">
<ul style="list-style-type: none; padding-left: 0; color: #0057e7;">
En este paso, profundizar√© en la estacionalidad y las tendencias en el comportamiento de compra de los clientes, lo que puede ofrecer valiosas ideas para adaptar las estrategias de marketing y mejorar la satisfacci√≥n del cliente. Aqu√≠ est√°n las caracter√≠sticas que estoy buscando introducir:
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> Gasto_Mensual_Promedio:</strong> Este es el monto promedio que un cliente gasta mensualmente. Nos ayuda a evaluar el h√°bito de gasto general de cada cliente. Un promedio m√°s alto indica un cliente que gasta m√°s, lo que podr√≠a mostrar inter√©s en productos premium, mientras que un promedio m√°s bajo podr√≠a indicar un cliente m√°s consciente de su presupuesto.</li>
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> Desviaci√≥n_Estandar_Gasto_Mensual:</strong> Esta caracter√≠stica indica la variabilidad en el gasto mensual de un cliente. Un valor m√°s alto se√±ala que el gasto del cliente fluct√∫a significativamente mes a mes, lo que podr√≠a indicar compras grandes y espor√°dicas. En contraste, un valor m√°s bajo sugiere h√°bitos de gasto m√°s estables y consistentes. Comprender esta variabilidad puede ayudar a crear promociones o descuentos personalizados durante los per√≠odos en que se espera que gasten m√°s.</li>
    <li style="padding-left: 20px;"><span style="color: #003366;">‚Ä¢</span> <strong> Tendencia_Gasto:</strong> Esto refleja la tendencia en el gasto de un cliente a lo largo del tiempo, calculada como la pendiente de la l√≠nea de tendencia lineal ajustada a sus datos de gasto. Un valor positivo indica una tendencia creciente en el gasto, lo que podr√≠a se√±alar una creciente lealtad o satisfacci√≥n. Por otro lado, una tendencia negativa podr√≠a se√±alar un inter√©s o satisfacci√≥n decreciente, resaltando la necesidad de estrategias de reenganche. Un valor cercano a cero indica h√°bitos de gasto estables. Reconocer estas tendencias puede ayudar a desarrollar estrategias para mantener o alterar los patrones de gasto de los clientes, mejorando la efectividad de las campa√±as de marketing.</li>

Al incorporar estas percepciones detalladas en nuestro modelo de segmentaci√≥n de clientes, podemos crear grupos de clientes m√°s precisos y accionables, facilitando el desarrollo de estrategias de marketing y promociones altamente dirigidas.

</div>

```python
# Extrayendo el mes y a√±o de InvoiceDate
# ==================================================================================
df['Anio'] = df['InvoiceDate'].dt.year
df['Mes'] = df['InvoiceDate'].dt.month

# Calcular el gasto mensual de cada cliente
# ==================================================================================
gasto_mensual = df.groupby(by = ['CustomerID','Anio','Mes'])['total_gastos'].sum().reset_index()

# Calcular Patrones de Compra Estacionales: Estamos utilizando la frecuencia mensual como un proxy para los patrones de compra estacionales.
# ==================================================================================
patrones_compras_estacionales = gasto_mensual.groupby(by = 'CustomerID')['total_gastos'].agg(['mean','std']).reset_index()
patrones_compras_estacionales.rename(columns = {'mean':'Promedio_gasto_mensual','std':'Desviacion_gasto_mensual'},inplace = True)

# Sustituir los valores NaN en Desviacion_gasto_mensual por 0, lo que implica que no hay variabilidad para los clientes con un mes de transacci√≥n √∫nico.
# ==================================================================================
patrones_compras_estacionales['Desviacion_gasto_mensual'].fillna(0,inplace = True)

# Calcular Tendencias en el Gasto: Estamos utilizando la pendiente de la l√≠nea de tendencia lineal ajustada al gasto del cliente 
# a lo largo del tiempo como un indicador de las tendencias de gasto
# ==================================================================================
def calcular_tendencia(data_tendencia):
    from scipy.stats import linregress
    #Si hay m√°s de un punto de datos, calculamos la tendencia utilizando regresi√≥n lineal.
    if len(data_tendencia) > 1:
        x = np.arange(len(data_tendencia))
        slope,_,_,_,_ = linregress(x,data_tendencia)
        return slope 
    else:
        return 0
    
# Aplicar la funci√≥n calcular_tendencia para encontrar la tendencia de gasto de cada cliente
# ==================================================================================
tendencia_gastos = gasto_mensual.groupby(by = 'CustomerID')['total_gastos'].apply(calcular_tendencia).reset_index()
tendencia_gastos.rename(columns = {'total_gastos':'Tendencia_gastos'},inplace = True)

# Ahora uniendo las nuevas variables al cliente_data 
# ==================================================================================
cliente_data = pd.merge(cliente_data,patrones_compras_estacionales,on = 'CustomerID')
cliente_data = pd.merge(cliente_data,tendencia_gastos,on = 'CustomerID')

# Mostrar resultado de la nuevas variables
# ==================================================================================
cliente_data.head(3)
```

Los valores at√≠picos son puntos de datos que son significativamente diferentes de la mayor√≠a de los otros puntos en el conjunto. Estos puntos pueden distorsionar los resultados de nuestro an√°lisis, especialmente en el clustering k-means, donde pueden influir significativamente en la posici√≥n de los centroides de los cl√∫steres. Por lo tanto, es esencial identificar y tratar estos valores at√≠picos de manera adecuada para lograr resultados de clustering m√°s precisos y significativos.

Dada la naturaleza multidimensional de los datos, ser√≠a prudente utilizar algoritmos que puedan detectar valores at√≠picos en espacios multidimensionales. Voy a utilizar el algoritmo Isolation Forest para esta tarea. Este algoritmo funciona bien para datos multidimensionales y es computacionalmente eficiente. A√≠sla las observaciones seleccionando aleatoriamente una caracter√≠stica y luego eligiendo aleatoriamente un valor de divisi√≥n entre los valores m√°ximo y m√≠nimo de la caracter√≠stica seleccionada.

</ul>

```python
# Encuentra los valores at√≠picos para cada columna
# ==============================================================================================================================
outliers = ((cliente_data[variables_numericas] < (Q1-1.5*IQR)) | (cliente_data[variables_numericas] > (Q3+1.5*IQR)))
outliers_count = outliers.sum()/len(cliente_data[variables_numericas])
print('Porcentaje de outliers por cada variable: \n')
print(outliers_count)
```

```python
# Inicializando el modelo IsolationForest con un par√°metro de contaminaci√≥n de 0.05
# ==================================================================================
model = IForest(contamination=0.05,random_state=42)

# Train IForest detector
# ==================================================================================
model.fit(cliente_data.iloc[:,1:].to_numpy())

# Obteniendo la predicci√≥n del cliente_data
# ==================================================================================
cliente_data['Is_outlier'] = model.predict(cliente_data.iloc[:,1:].to_numpy())  # binary labels (0: inliers, 1: outliers)

# Mostrar las primeras filas del marco de datos cliente_data
# ==================================================================================
cliente_data.head()
```
